```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = [1,2,3,4,5],y = [3,4,5,6,7],mode = "markers"))
##add_trace is method used to add one or more traces to a plot. traces represent individual sets of data and other aspect of plot that can be visualization on the same plot.
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = [1,2,3,4,5],y = [3,4,5,6,7],mode = "lines"))
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill,y = tips.tip,mode = 'markers'))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill,y = tips.tip,mode = "markers",marker_size = 12*tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill,y = tips.tip,mode = 'lines'))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1,2,3,4,5],y = [3,4,5,6,7]))
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x = tips.total_bill,y = tips.tip))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = tips.total_bill,y = tips.tip,mode = "markers",z = tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = [1,2,3,4,1],y = [4,5,6,7,2],mode = "lines",z = [1,2,3,4,5]))
```
