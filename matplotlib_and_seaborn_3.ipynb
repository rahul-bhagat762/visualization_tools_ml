```python
##piechart
# using matplot lib
langs = ['c','python','java']
students = [20,100,40]
plt.pie(students,labels=langs,autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
##piechart
# using matplot lib
my_explor = (0.1,0.9,0.2)
langs = ['c','python','java']
students = [20,100,40]
plt.pie(students,labels=langs,autopct='%1.1f%%',explode=my_explor,shadow=True)
plt.title("Pie Chart")
plt.show()
```
