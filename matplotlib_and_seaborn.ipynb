```python
plt.figure() ##make a page
plt.subplot(1,2,1) #divide a page into 2 subparts and acces first part 
plt.plot(x,np.sin(x)) ##this line is use to plot a graph in first part
plt.subplot(1,2,2) #divide a page into 2 subparts and acces second part 
plt.plot(x,np.cos(x)) #this line is use to plot a graph in second part
plt.show() ##this line is used to show as a form of image of both canvas in a single image 
```

```python
plt.figure() ##make a page
plt.subplot(2,2,1) #divide 2 pages into 2 subparts and acces first part 
plt.plot(x,np.sin(x)) ##this line is use to plot a graph in first part
plt.subplot(2,2,2) #divide 2 pages into 2 subparts and acces second part 
plt.plot(x,np.cos(x)) #this line is use to plot a graph in second part
plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.subplot(2,2,4)
plt.plot(x,np.sin(x+1))
plt.show() ##this line is used to show as a form of image of both canvas in a single image 
```

```python
plt.figure()
plt.subplot(2,1,1) ##in this line our page is divide into 2 but rows are 2 and col is 1 and we want to access 1st
plt.plot(x,np.sin(x))

plt.figure()
plt.subplot(2,1,2) ##in this line our page is divide into 2 but rows are 2 and col is 1 and we want to access 2nd
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x-0),color = 'black')
plt.plot(x,np.sin(x-1),color = 'g')
plt.plot(x,np.sin(x-2),color = (1.0,0.5,0)) ##rgb format
```

```python
plt.plot(x,np.sin(x-0),color = 'black')
plt.ylim(-1.5,1.5) ##when you want to change your limits of y axis(from some value to some value)
plt.xlim(0,15) ##when you want to change your limits of x axis(from some value to some value)
plt.show()
```

```python
# when we want to use this extra space means expand our graph and use this extra space
plt.plot(x,np.sin(x-0),color = 'black')
plt.ylim(-1.5,1.5) ##when you want to change your limits of y axis(from some value to some value)
plt.xlim(0,15) ##when you want to change your limits of x axis(from some value to some value)
plt.axis('equal')
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color = 'black')
plt.ylim(-1.5,1.5) 
plt.xlim(0,15) 
plt.axis('equal')
plt.grid() ##this grid shows the actual grid on the graph which shoes which line is exists on which number
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color = 'black')
plt.ylim(-1.5,1.5) 
plt.xlim(0,15) 
plt.axis('equal')
plt.grid() ##this grid shows the actual grid on the graph which shoes which line is exists on which number
plt.title("This is a practice graph") #this line is used to add title on the graph
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color = 'black')
plt.ylim(-1.5,1.5) 
plt.xlim(0,15) 
plt.axis('equal')
plt.grid() ##this grid shows the actual grid on the graph which shoes which line is exists on which number
plt.title("This is a practice graph") #this line is used to add title on the graph
plt.xlabel("this is x axis --->")
plt.ylabel("This is y axis --->")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize = (6,2))
plt.scatter(x,y,c = 'maroon')
plt.ylabel("y")
plt.xlabel("x")
plt.title("This is a scater plot")
plt.axis('equal')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (6,2))
plt.scatter(x,y,c = 'r',alpha= 0.9) ##alpha is used to control the intensity of the graph(which is vary from (0,1))
plt.ylabel("y")
plt.xlabel("x")
plt.title("This is a scater plot")
plt.axis('equal')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (6,2))
plt.scatter(x,y,c = 'r',marker='*') ##marker parameter is used to change this display formation means change its marking points
plt.ylabel("y")
plt.xlabel("x")
plt.title("This is a scater plot")
plt.axis('equal')
plt.grid()
plt.show()
```

```python
##for ,categorical,discreate data we will make bar plot
x = ['a','b','c','d']
y = [3,4,5,1]
```

```python
plt.figure(figsize=(5,6))
plt.bar(x,y)
plt.title("This is a bar plot")
plt.xlabel("x-->")
plt.ylabel("y-->")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.barh(x,y) ##this is used to show the graph in horizontal manner
plt.title("This is a bar plot")
plt.xlabel("x-->")
plt.ylabel("y-->")
plt.grid()
plt.show()
```

```python
##3 d plot
# data = [1,2,3,3,4,5,6,4,4,8,9,7,8]
x=np.random.randn(20)
y=np.random.randn(20)
z = np.random.randn(20)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,c='r',alpha=0.9)
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.show()
```
