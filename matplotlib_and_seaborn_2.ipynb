```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Bank_churn.csv')
```

```python
df.head(1)
```

```python
##Give Insights 
df.plot.scatter(x = 'Age',y = 'Balance')
plt.show()
##insights 
# Maximum concentration >> 25-50
# maximum == age == 40
# min == 0, it is present in all of the group
```

```python
##Age Distribution
plt.title("Age Distribution")
plt.hist(df['Age'],color='r')
plt.show()

##Balance Distribution
plt.title("Balance Distribution")
plt.hist(df['Balance'],color='b')
plt.show()
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill']) ##this line is called as karnel density stimiter
```

```python
sns.distplot(tips['total_bill'],kde=False) #when you dont want to see this line then you can simply use kde = false
```

```python
sns.distplot(tips['total_bill'],kde=False,bins = 30)#when you want to increase the number of bins then use this bins attribute
```

```python
##cater plot
sns.scatterplot(x = tips['total_bill'],y=tips.tip,data = tips,hue= 'smoker')
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,hue= 'smoker')
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,style= 'size')
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,style= 'size',hue='time')
```

```python
##bar plot or categorical plot
sns.catplot(x = 'day',y= 'total_bill',data= tips,hue='smoker')
```

```python
sns.barplot(x = 'day',y = 'total_bill',data = tips,hue='smoker')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips)
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,hue='smoker')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind='reg')
```
